```python
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3],[4,5,1])
   ## (1,4),(1,5)(1,1)(2,4)....thats like 
plt.title('information')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
x=[1,2,3]
y=[4,5,6]
x2=[7,8,9]
y2=[10,11,12]
plt.plot(x,y,'r',label='line',linewidth=5)
plt.plot(x2,y2,'g',label='line two',linewidth=8)
plt.legend()  ## when legend function calls then only labels are visible
plt.title('graph')
plt.xlabel('x axis')
plt.ylabel('y axis') 
plt.grid('black')
plt.show()
```

*bar graph**

```python
plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,20],label="BMW",color='r',width=.1)
plt.bar([0.75,1.75,2.75,3.75,4.75],[80,20,20,50,60],label="AUDI",color='g',width=.8)
plt.legend()
plt.xlabel('days')
plt.ylabel('distance')
plt.title('information')
plt.show()
```

***histogram***

```python
###bin refers to the range of values
#age group----15-20,20-25,25-30
age=[18,24,23,25,21,17,18,30,21]
bins=[0,5,10,15,20,25,30,35]
plt.hist(age,bins,histtype='bar')  ##histtype='step'or 'bar'
plt.show()
```

#scatter plot

```python
## collection of variables each variable has x and y cooridinates that determines the position of x and y
x=[2,2.5,3,3.5,4,4.5,5.5]
y=[8,8.5,9,9.5,10,10.1,10.2]
y2=[2,2.5,3,3.5,4,4.5,5.5]
x2=[8,8.5,9,9.5,10,10.1,10.2]
plt.scatter(x,y,color='r',label='line 1')
plt.scatter(x2,y2,color='g',label='line 2')
plt.legend()
plt.xlabel("income")
plt.ylabel("savings")
plt.title('scatter plot')
plt.plot()
```

#3d plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris=pd.read_csv("Iris.csv")
```

```python
iris
```

```python
for Species,irissubset in iris.groupby('Species'):
  plt.scatter(irissubset['PetalLengthCm'],irissubset['PetalWidthCm'],label=Species)
  plt.legend()
```

```python
#from mpl_toolkits.mplot3d import axes3d     it is no need
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
for Species,irissubset in iris.groupby('Species'):
  ax.scatter(irissubset['PetalLengthCm'],irissubset['PetalWidthCm'],irissubset['SepalWidthCm'],label=Species)
  ax.set_xlabel('PetalLength')
  ax.set_ylabel('PetalWidth')
  ax.set_zlabel('SepalWidth')
  ax.legend()
```
